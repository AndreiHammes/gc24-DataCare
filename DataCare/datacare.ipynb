### S02T03 - Identificar e tratar dados faltantes

```python
import pandas as pd
```

```python
df = pd.read_csv("depression_dataset.csv")
df.head(15)
```

```python
df.isnull().sum()
```

```python
df[['CGPA', 'Study Satisfaction', 'Academic Pressure', 'Work Pressure', 'Job Satisfaction']] = df[['CGPA', 'Study Satisfaction', 'Academic Pressure', 'Work Pressure', 'Job Satisfaction']].fillna(0)
df['Profession'] = df['Profession'].fillna('None')
df.isnull().sum()
```

```python
df.to_csv('depression_dataset_treated.csv')
```
